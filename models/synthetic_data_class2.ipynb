```python
import random

import numpy as np
import cv2
```

```python
# Plain Black background image
# Let's create a 100 x 100 matrix with all pixels in black color
#img = np.zeros((100,100,3),np.uint8)
img = np.zeros([100,100,3],dtype=np.uint8)
# img.fill(255) # or img[:] = 255
img[:] = 0

#save our image (numpy array) as a "jpg" image
cv2.imwrite("../images/class1.jpg",img)

# Load created image - see plot
import matplotlib.pyplot as plt
from matplotlib import image as img
image = img.imread('../images/class1.jpg')
print(image.shape)
plt.imshow(image)

dytype = type(image)
print('data type',dytype)
```

```python
# Parameters to draw a circle
center = (20, 40)
radius = 15

# Draws a circle
img_c = cv2.circle(image, center, radius, color=(255,0,0), thickness=1, lineType=8, shift=0)

#Parameters:

#img (CvArr) – Image where the circle is drawn
#center (CvPoint) – Center of the circle
#radius (int) – Radius of the circle
#color (CvScalar) – Circle color
#thickness (int) – Thickness of the circle outline if positive, otherwise this indicates that a filled circle is to be drawn
#lineType (int) – Type of the circle boundary, see Line description
#shift (int) – Number of fractional bits in the center coordinates and radius value
```

```python
print(img_c.shape)
itype = type(img_c)
print(itype)

# Display image

plt.imshow(img_c)
```

```python
# Make a function to draw black background image class1

def draw_class1():
    img = np.zeros([100,100,3],dtype=np.uint8)
    img[:] = 0
    return img


for number in range(5):

    f = plt.figure()

    imgclass1 = draw_class1()
    plt.imshow(imgclass1)
    f.savefig("../data/class1/class1_img" + str(number) + '.jpg')
```

```python
# Make a function to draw a circle image class 2

def draw_class2(img,x,y,radius):
    #  x,y for position/center of circle
    img_c = cv2.circle(img, (x,y), radius, color=(255,0,0), thickness=0, lineType=8,shift=0)
    return img_c

# x,y any number b/w 20 and 80
# radius any number b/w 5 and 15


for number in range(5):

    f = plt.figure()

    x = random.randint(20,80)
    y = random.randint(20,80)
    radius = random.randint(5,15)

    print('center_coordinates',x,y,'radius',radius)
    imgclass1 = draw_class1()
    img_class2 = draw_class2(imgclass1,x, y, radius)
    plt.imshow(img_class2)
    f.savefig("../data/class2/class2_img" + str(number) + '.jpg')

```
